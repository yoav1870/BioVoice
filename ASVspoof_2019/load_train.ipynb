```python
from datasets import load_dataset

ds = load_dataset("Bisher/ASVspoof_2019_LA", split="train")
print(ds)
```

```python
import pandas as pd

df = ds.to_pandas()

num_speakers = df["speaker_id"].nunique()
print("Number of unique speakers:", num_speakers)
```

```python
label_counts = df["key"].value_counts()
print(label_counts)

print("\nPercentage:")
print(df["key"].value_counts(normalize=True) * 100)
```

```python
samples_per_speaker = df.groupby("speaker_id").size()

print("Average samples per speaker:", samples_per_speaker.mean())
print("Min samples:", samples_per_speaker.min())
print("Max samples:", samples_per_speaker.max())
```

```python
system_counts = df["system_id"].value_counts()

print("Number of spoof systems:", len(system_counts))
print(system_counts.head())
```

```python
speaker_label_distribution = (
    df.groupby(["speaker_id", "key"]).size().unstack(fill_value=0)
)

print(speaker_label_distribution.head())
```
